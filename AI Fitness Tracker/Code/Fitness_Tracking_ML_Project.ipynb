```python
# making a gernal fuction for reading the data and getting acc_file and gyr_file

data_path = 'D:\Projects\AI Fitness Tracker\MetaMotion\MetaMotion\\'
files = glob(data_path + '*.csv')

def read_data(files):
    
    
    acc_df = pd.DataFrame()
    acc_count = 1
    gyr_df = pd.DataFrame()
    gyr_count = 1

    for f in files:
        participant = f.split("-")[0].replace(data_path,'')
        label = f.split("-")[1]
        category = f.split('-')[2].replace('_MetaWear_2019','')
        category = remove_number_from_string(category)
        df = pd.read_csv(f)
        df["Participant"] = participant
        df["Label"] = label
        df["Category"] = category

        if "Accelerometer" in f:
            df["Index"] = acc_count
            acc_count += 1
            acc_df = pd.concat([acc_df, df])
        else:
            df["Index"] = gyr_count
            gyr_count += 1
            gyr_df = pd.concat([gyr_df, df])
            
    acc_df.index = pd.to_datetime(acc_df["epoch (ms)"], unit = 'ms')
    del acc_df["epoch (ms)"]
    del acc_df["time (01:00)"]
    del acc_df["elapsed (s)"]
    gyr_df.index = pd.to_datetime(gyr_df["epoch (ms)"], unit = 'ms')
    del gyr_df["epoch (ms)"]
    del gyr_df["time (01:00)"]
    del gyr_df["elapsed (s)"]
        
    return acc_df, gyr_df

acc_df, gyr_df = read_data(files)
```

```python

combined_df= pd.concat([acc_df, gyr_df], axis = 1)
# Remove duplicate columns from the concatenated DataFrame
combined_df = combined_df.loc[:, ~combined_df.columns.duplicated()]
combined_df
```

```python
# Rearrange the columns
combined_df = combined_df[['x-axis (g)', 'y-axis (g)', 'z-axis (g)', 'x-axis (deg/s)', 'y-axis (deg/s)', 'z-axis (deg/s)', 'Participant', 'Label', 'Category', 'Index']]
combined_df
```

```python
# Define a dictionary for renaming columns
column_rename_dict = {
    'x-axis (g)': 'acc_x',
    'y-axis (g)': 'acc_y',
    'z-axis (g)': 'acc_z',
    'x-axis (deg/s)': 'gyr_x',
    'y-axis (deg/s)': 'gyr_y',
    'z-axis (deg/s)': 'gyr_z',
    'Participant': 'participant',
    'Label': 'label',
    'Category': 'category',
    'Index': 'set'
}

# Use the rename method to apply the column name changes
combined_df.rename(columns=column_rename_dict, inplace=True)
```

```python
sampling = {
    'acc_x': 'mean',   
    'acc_y': 'mean',
    'acc_z': 'mean',
    'gyr_x': 'mean',
    'gyr_y': 'mean',
    'gyr_z': 'mean',
    'participant': 'last',
    'label': 'last',
    'category': 'last',
    'set': 'last'
}
days = [g for n, g in combined_df.groupby(pd.Grouper(freq='D'))]
```
